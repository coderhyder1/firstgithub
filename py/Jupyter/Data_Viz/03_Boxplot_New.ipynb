```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
kashti.head()
```

```python
sns.boxplot(x="survived",
            y="age",
            data=kashti)
```

```python
p1 = sns.boxplot(x="survived",
            y="age",
                 showmeans=True,
            data=kashti)
```

```python
p1 = sns.boxplot(x="survived",
            y="age",
                 showmeans=True,
                 meanprops= {"marker": "+",
                            "markersize":"12",
                            "markeredgecolor":"red"},
            data=kashti)
#Show labels
plt.xlabel("How many survived?", size = 14),
plt.ylabel("Age (years)", size = 12),
plt.title("box plot of kitne doobe?", size= 12)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

p1 = sns.boxplot(x="survived",
                 y="age",
                 showmeans=True,
                 meanprops={"marker": "+",
                            "markersize": "12",
                            "markeredgecolor": "red"},
                 data=kashti)

# Show labels without using parentheses
plt.xlabel("How many survived?", size=14)
plt.ylabel("Age (years)", size=12)
plt.title("Box plot of kitne doobe?", size=12, weight= "bold")
plt.show()
```
