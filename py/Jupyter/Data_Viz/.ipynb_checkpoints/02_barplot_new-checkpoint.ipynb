```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
kashti = sns.load_dataset("titanic")
kashti

#draw a lineplot
sns.barplot(x= "sex", y="alone", hue= "who", data= kashti, order = ["female", "male"])
#hue is for adding colors
#order is for changing the orders, it is to be placed in [] brackets and each element in "".
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
kashti = sns.load_dataset("titanic")
kashti

#draw a lineplot
sns.barplot(x= "sex", y="alone", hue= "who", data= kashti, order = ["female", "male"],
            color = "salmon")
#hue is for adding colors
#order is for changing the orders, it is to be placed in [] brackets and each element in "".
#color for the colors of the graph
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset
kashti = sns.load_dataset("titanic")
kashti

#draw a lineplot
sns.barplot(x= "sex", y="alone", hue= "who", data= kashti, order = ["female", "male"],
            color = "salmon", errorbar = None, palette="BrBG")
#hue is for adding colors
#order is for changing the orders, it is to be placed in [] brackets and each element in "".
#color for the colors of the graph
#errorbar = None is to remove the error bar.

plt.show()
```

## Horizontal Plot
- we just have to put the numeric variable on x axis and catagorical variable on y axis.

```python
#import libraries
import seaborn as sns
import numpy
import matplotlib.pyplot as plt

#Load dataset
kashti = sns.load_dataset("titanic")
kashti

#draw a lineplot
sns.barplot(x= "fare", y="class", hue= "sex", data= kashti, estimator=mean, saturation=.2)
#hue is for adding colors
#saturation for intensity of colors

plt.show()
```

```python
#import libraries
import seaborn as sns
import numpy
import matplotlib.pyplot as plt

#Load dataset
kashti = sns.load_dataset("titanic")

#draw a lineplot
sns.barplot(x= "class", y="fare", data= kashti,
           linewidth= 3, facecolor= (0,1,0,.5),
           errcolor="0", edgecolor="0")
#linewidth is width of the border of bar
#facecolor is RGB and Saturation. we can adjust the color of the figure by adjusting these values. 
#errcolor is for adjustment of color of the error line.
#edgecolor is the color of the bar border.
plt.show() 
```
